```python
def plot_eis_for_first_battery(file_path):
    font = FontProperties(family='Arial', size=11)
#     title = file_path.replace('.txt', '')
    parts = file_path.split('_')  # 分割字符串
    desired_parts = [parts[0], parts[1], parts[2]]  # 选择需要的部分
    title = '_'.join(desired_parts)
    title =title.replace('.txt', '')
    lines = []
    with open(file_path, 'r') as file:
        for i, line in enumerate(file, start=1):
            if i >= 1 and i <= 180 and i % 10 == 0:  # Selecting only lines that are multiples of 10
                lines.append(line.strip())

    # Plotting the EIS graph
    plt.figure(figsize=(10, 8))
    colors = plt.cm.viridis(np.linspace(0, 1, len(lines)))  # Color spectrum for lines

    for i, line in enumerate(lines):
        parts = line.split()
        x_values = [float(val) for val in parts[:60]]
        y_values = [float(val) for val in parts[60:]]
        plt.scatter(x_values, y_values, color=colors[i], label=f'cycle {i*10 + 1}')
    
    plt.title(title, fontdict={'size': 15, 'family': 'Arial'})
    plt.xlabel('Re[Z](ohm)', fontdict={'size': 20, 'family': 'Arial'})
    plt.ylabel('-Im[Z](ohm)', fontdict={'size': 20, 'family': 'Arial'})
    
    plt.legend(prop=font, facecolor='white', frameon=False,loc='best')
#     plt.legend(prop=font, facecolor='white', frameon=False)
    plt.grid(False)
    # 设置横坐标和纵坐标刻度的字体大小
    plt.xticks(fontsize=15,fontproperties=font)  # 设置横坐标刻度的字体大小为 15
    plt.yticks(fontsize=15,fontproperties=font)  # 设置纵坐标刻度的字体大小为 15
    plt.savefig( f"EIS图_{title}.pdf", format='pdf', dpi=600, bbox_inches='tight')
    plt.savefig( f"EIS图_{title}.jpg", format='jpg', dpi=1000, bbox_inches='tight')
    plt.show()
```

```python
##色度EIS图
from matplotlib.cm import ScalarMappable

def plot_eis_for_first_battery(file_path):
    font = FontProperties(family='Arial', size=17)
    title = file_path.split('_')
    title = '_'.join(title[:3]).replace('.txt', '')
    
    with open(file_path, 'r') as file:
        lines = [line.strip() for i, line in enumerate(file, start=1) if i >= 1 and i <= 180 and i % 10 == 0]

    plt.figure(figsize=(10, 8))
    colors = plt.cm.viridis(np.linspace(0, 1, len(lines)))

    for i, line in enumerate(lines):
        parts = line.split()
        x_values = [float(val) for val in parts[:60]]
        y_values = [float(val) for val in parts[60:]]
        plt.scatter(x_values, y_values, color=colors[i])

    plt.title(title, fontdict={'size': 20, 'family': 'Arial'})
    plt.xlabel('Re[Z](ohm)', fontdict={'size': 20, 'family': 'Arial'})
    plt.ylabel('-Im[Z](ohm)', fontdict={'size': 20, 'family': 'Arial'})
    plt.grid(False)
    plt.xticks(fontsize=17, fontproperties=font)
    plt.yticks(fontsize=17, fontproperties=font)
    plt.axis('equal')
    # Create a color bar with a ScalarMappable object
    sm = ScalarMappable(cmap=plt.cm.viridis, norm=plt.Normalize(vmin=1, vmax=180))
    sm.set_array([])  # You can set an empty list as the initial array
    cbar = plt.colorbar(sm, ticks=np.linspace(1, 180, 2))  # Adjust the ticks according to your preference
    cbar.ax.set_yticklabels(['Fresh', 'Aged'],fontsize=17, fontproperties=font)  # Set the tick labels
#     cbar.set_label('Cycle Number', fontproperties=font)

    plt.savefig(f"色度EIS图新_{title}.pdf", format='pdf', dpi=600, bbox_inches='tight')
    plt.savefig(f"色度EIS图新_{title}.jpg", format='jpg', dpi=1000, bbox_inches='tight')
    plt.show()

```
