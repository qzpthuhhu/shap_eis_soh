```python
# Replace this with your actual file path
font = FontProperties(family='Arial', size=15)
folder_path = "C:/Users/10193/SHAP-tutorial/SOH1/"

# Use glob to find all CSV files in the folder
csv_files = glob.glob(os.path.join(folder_path, "*.csv"))
font = FontProperties(family='Arial', size=14)

# 
plt.figure(figsize=(15, 5))
legend_labels = ['B1','B2','B3','B4','B5','B6','B7','B8']
for i, csv_file in enumerate(csv_files):
    df = pd.read_csv(csv_file)
    cycle_number = df.iloc[:, 0].values
    capacity = df.iloc[:, 1].values
    source_file_name = csv_file.split("\\")[-1].replace(".csv", "")
    source_file_name = source_file_name.replace("Data_Capacity_", "")
    plt.plot(cycle_number, capacity, label=source_file_name, alpha=1,lw=4)


plt.xticks(fontsize=15,fontproperties=font)  #  15
plt.yticks(fontsize=15,fontproperties=font)  #  15
# 
plt.legend(legend_labels, prop=font, loc='upper right', bbox_to_anchor=(1.0, 1.0), facecolor='white', frameon=False)

# 
plt.xlabel("Cycle Number", fontdict={'size': 20, 'family': 'Arial'})
plt.ylabel("Capacity (mA.h)", fontdict={'size': 20, 'family': 'Arial'})

# # PNGSVG
output_filename = "Capacity Cycle Number"

plt.savefig( f"{output_filename}.pdf", format='pdf', dpi=600, bbox_inches='tight')
plt.savefig( f"{output_filename}.jpg", format='jpg', dpi=1000, bbox_inches='tight')

# 
plt.show()
```
